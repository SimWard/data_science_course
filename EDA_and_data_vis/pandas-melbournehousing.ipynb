```python
# if lower than 25 percentile = small, if lower than 75 = medium, if greater or equal to 75 = large

def recode_landsize(s):
    if np.isnan(s):
        i = 'UNKNOWN'
    elif s < 224:
        i = 'SMALL'
    elif s < 521:
        i = 'MEDIUM'
    else:
        i = 'LARGE'
    
    return i
```

```python
df['LandSizeCategory'] = df['Landsize'].apply(recode_landsize)
```
