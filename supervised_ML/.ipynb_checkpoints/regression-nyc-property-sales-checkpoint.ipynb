```python
# Data from https://www.kaggle.com/new-york-city/nyc-property-sales

import pandas as pd
import seaborn as sns
import datetime as dt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
sales = pd.read_csv('nyc-rolling-sales.csv')
```

```python
sales = sales.loc[(sales['LAND SQUARE FEET']  != ' -  ') & (sales['GROSS SQUARE FEET'] != ' -  ') & (sales['SALE PRICE']  != ' -  ')]
```

```python
sales['SALE DATE'] = pd.to_datetime(sales['SALE DATE'])
sales['sale_year'] = sales['SALE DATE'].dt.year
sales['sale_month'] = sales['SALE DATE'].dt.month
```

```python
sales['SALE PRICE']
```

```python
sales['LAND SQUARE FEET'] = sales['LAND SQUARE FEET'].astype(int)
sales['GROSS SQUARE FEET'] = sales['GROSS SQUARE FEET'].astype(int)
sales['SALE PRICE'] = sales['SALE PRICE'].astype(float)
```

```python
# let's just make life easier for ourselves and remove weird columns
sales_predict = sales[['YEAR BUILT', 'ZIP CODE', 'LAND SQUARE FEET', 'sale_year', 'sale_month', 'SALE PRICE']].copy()
```

```python
sns.pairplot(sales_predict)
```

```python
sns.heatmap(sales_predict.corr(),annot=True)
```

```python
def prepare_data(df, target_column, test_size=0.3, random_state=42):
    # This assumes all other columns except target_column is features
    
    features = [i for i in df.columns if i != target_column]
    
    X = df[features]
    y = df[target_column]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        test_size=test_size, 
                                                        random_state=random_state)
    
    return X_train, X_test, y_train, y_test

def train_model(X_train, y_train, model):
    return model.fit(X_train, y_train)

def produce_model_predictions(fitted_model, X_test, y_test):
    prediction = fitted_model.predict(X_test)
    comparison = pd.DataFrame({
                               "prediction": prediction,
                               "actual": y_test
                             })
    return comparison

def return_model_mse(df):
    # df should have two columns, the prediction and actuals
    df["error"] = df["prediction"] - df["actual"]
    df["squared_error"] = df["error"]**2

    mse = df['squared_error'].mean()
    
    return mse
```

```python
X_train, X_test, y_train, y_test = prepare_data(sales_predict, "SALE PRICE", test_size=0.3, random_state=42)
```

```python
reg = train_model(X_train, y_train, linear_model.LinearRegression())
```

```python
comparison = produce_model_predictions(reg, X_test, y_test)
```

```python
comparison['prediction'] = np.log(comparison['prediction'])
comparison['actual'] = np.log(comparison['actual'])
```

```python
sns.scatterplot(data=comparison)
```
