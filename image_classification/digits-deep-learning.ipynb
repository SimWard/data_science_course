```python
import pandas as pd
import sklearn

from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split

from tensorflow import keras
from tensorflow.keras import layers

dat = datasets.load_digits()

digits = pd.DataFrame(dat['data'])

digits['digit'] = dat['target']
digits
```

```python
y = digits['digit']
X = digits.drop('digit', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
input_shape = X.shape[1]

model = keras.Sequential(
        [
            keras.Input(shape=input_shape),
            layers.Dense(10, activation='relu'),
            layers.Dense(64, activation='relu'),
            layers.Dense(24, activation='relu'),
            layers.Dense(24, activation='relu'),
            layers.Dense(10, activation='softmax') # This states that the highest probability classification gets chosen as the correct
        ])
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
batch_size = 128
epochs = 1000

#callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

model.fit(X_train, y_train, batch_size=batch_size, validation_split=0.25)
```
