```python
import pandas as pd

# read the data
houses = pd.read_csv("./train.csv")

# drop the id columns
houses = houses.drop(columns="Id")

houses
```

```python
Xy = houses.dropna(axis=1)
```

```python
cat_list = []

for col, v in dict(zip(Xy.columns.tolist(), (Xy.dtypes == "object").tolist())).items():
    if v:
        cat_list.append(col)

# these are our first 10 categorical columns
cat_list[:10]
```

```python
from sklearn import preprocessing

# this is a categorical data
cat_data = Xy[cat_list]

# one hot encoding
ohc = preprocessing.OneHotEncoder(handle_unknown='ignore')

X_cat = ohc.fit_transform(cat_data)
```

```python
Xy_clean = pd.concat(
    [pd.DataFrame(X_cat.todense())
     , Xy.drop(columns=cat_list)], axis=1
)
```
