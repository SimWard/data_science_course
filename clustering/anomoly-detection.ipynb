## Multivariate outlier detection

This can be useful when looking at fraud, unusual customers, etc but here we'll just use it for fraud.

```python
from sklearn import datasets
from sklearn import neighbors
import pandas as pd
import seaborn as sns

dat = datasets.load_iris()

iris = pd.DataFrame(data=dat['data'], columns=dat['feature_names'])
iris['species'] = dat['target']
iris
```

```python
X = iris.drop('species', axis=1).copy()
```

```python
sns.pairplot(data=iris)
```

```python
lof = neighbors.LocalOutlierFactor()
```

```python
outliers = lof.fit_predict(X)
outliers
```

```python
evaluate = X.copy()
evaluate['outliers'] = outliers
evaluate['outliers'] = evaluate['outliers'].astype('str')
```

```python
sns.pairplot(evaluate, hue='outliers')
```
