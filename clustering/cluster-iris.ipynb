## Cluster exploration with Iris

```python
from sklearn import datasets
from sklearn import cluster
import pandas as pd
import seaborn as sns

dat = datasets.load_iris()

iris = pd.DataFrame(data=dat['data'], columns=dat['feature_names'])
iris['species'] = dat['target']
iris
```

Classification doesn't use a target feature. It only has features

```python
sns.pairplot(iris.drop('species', axis=1))
```

```python
X = iris.drop('species', axis=1).copy()
```

```python
def visualise_model(classifier, X=X):
    cls.fit(X)
    X['labels'] = cls.labels_
    X['labels'] = X['labels'].astype('str')
    sns.pairplot(X, hue='labels')
```

#### KMeans

```python
cls = cluster.KMeans(n_clusters=3)
visualise_model(cls)
```

#### Mean shift

```python
cls = cluster.MeanShift()
visualise_model(cls)
```

#### Spectral Clustering

```python
cls = cluster.AgglomerativeClustering(n_clusters=3)
visualise_model(cls)
```

#### DBSCAN

```python
cls = cluster.DBSCAN()
visualise_model(cls)
```

#### Optics

```python
cls = cluster.OPTICS()
visualise_model(cls)
```

#### Birch

```python
cls = cluster.Birch(n_clusters=3)
visualise_model(cls)
```
