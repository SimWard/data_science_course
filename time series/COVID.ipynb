```python
# Univariate time series analysis - ignoring exogenous factors

# https://ourworldindata.org/coronavirus-source-data
```

```python
import pandas as pd

covid = pd.read_csv('./owid-covid-data.csv', )
covid.head()
```

```python
covid_uk = covid[covid['location'] == 'United Kingdom']
covid_uk = covid_uk.loc[:, ['date', 'new_cases']].copy()
covid_uk['date'] = pd.to_datetime(covid_uk['date'])
covid_uk = covid_uk.set_index('date')
```

```python
covid_uk.plot();
```

```python
# Rolling average
covid_uk['rolling_7_days'] = covid_uk['new_cases'].rolling(7, center=True).mean()
covid_uk.plot();
```

```python
# Seasonal decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

decomp = seasonal_decompose(covid_uk['new_cases'], freq=7)
decomp.plot()
plt.show()
```

```python
# Autocorrelation function

from statsmodels.graphics.tsaplots import plot_acf

plot_acf(covid_uk['new_cases'], lags=300)
plt.show()
```
